```python
# taking y separately to df11 (train data)
df10=df7.drop(['y'],axis=1)
df11=df7['y']
print(df10)
print(df11)
```

```python
# pca for train data
from sklearn.decomposition import PCA
pca=PCA(n_components=14)
df12=pca.fit_transform(df10)
df13=pd.DataFrame(df12)
print(df13)
print(df13.shape)
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
#x,y variable for train_test_split
x=df13
y=df11
print(x)
print(y)
```

```python
# model selection for train data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=123)
```

```python
# xgboost regressor for given train data 
from xgboost import XGBRegressor
xg=XGBRegressor(objective='reg:linear',learning_rate=0.1,n_jobs=-1)
xg.fit(x_train,y_train)
predicted1=xg.predict(x_train)
print(predicted1)


```

```python
predicted2=xg.predict(x_test)
predicted2
```

```python
# performance metrics
from sklearn.metrics import mean_squared_error,r2_score
mse1=mean_squared_error(y_train,predicted1)
rmse1=np.sqrt(mse1)
r1=r2_score(y_train,predicted1)
print(mse1,rmse1,r1)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse2=mean_squared_error(y_test,predicted2)
rmse2=np.sqrt(mse2)
r2=r2_score(y_test,predicted2)
print(mse2,rmse2,r2)
```

```python
# RandomForest regressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_jobs=-1)
rf.fit(x_train,y_train)
RF_predict1=rf.predict(x_train)
RF_predict2=rf.predict(x_test)
print(RF_predict1)
print(RF_predict2)

```

```python
#performance metrics for y_test
from sklearn.metrics import mean_squared_error,r2_score
RF_mse=mean_squared_error(y_test,RF_predict2)
RF_r2=r2_score(y_test,RF_predict2)
print(RF_mse)
print(RF_r2)
```

XGBoost has less mse value than RandomForest for test data so continuing with XGboostRegressor
